# **Breadth First Search**

```python
graph = {'A':['B','C','D'],
         'B':['E','F'],
         'C':['G','H'],
         'D':['I','J'],
         'E':['K','L'],
         'F':['L','M'],
         'G':['N'],
         'H':['O','P'],
         'I':['P','Q'],
         'J':['R'],
         'K':['S'],
         'L':['T'],
         'P':['U'],
         'S':[],
         'T':[],
         'M':[],
         'N':[],
         'O':[],
         'U':[],
         'Q':[],
         'R':[]}

def bfs_traversal(graph, start, goal):
    opened = [start]
    closed = []
    while opened:
        print(opened)
        node = opened.pop(0)
        if node is goal:
          closed.append(node)
          return closed
        else:
            opened = opened + [item for item in graph[node] if item not in opened and item not in closed]
            closed.append(node)
    return 'GOAL Not FOUND'


print("BFS")
bfsPath = bfs_traversal(graph,'A','M')
print(bfsPath)


```

# **Depth First Search**

```python

graph = {'A':['B','C','D'],
         'B':['E','F'],
         'C':['G','H'],
         'D':['I','J'],
         'E':['K','L'],
         'F':['L','M'],
         'G':['N'],
         'H':['O','P'],
         'I':['P','Q'],
         'J':['R'],
         'K':['S'],
         'L':['T'],
         'P':['U'],
         'S':[],
         'T':[],
         'M':[],
         'N':[],
         'O':[],
         'U':[],
         'Q':[],
         'R':[]}

def dfs_traversal(graph, start, goal):
    closed = []
    opened = [start]
    while opened:
        node = opened.pop(0)
        if node is goal:
          closed.append(node)
          return closed
        else:
          opened = [item for item in graph[node] if item not in opened and item not in closed] + opened
          closed.append(node)
    return 'GOAL Not FOUND'

print("DFS")
dfsPath=dfs_traversal(graph,'A','M')
print(dfsPath)
```
